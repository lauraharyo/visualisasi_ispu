## **Data preprocessing**

Gabung data ISPU selama tiga bulan menggunakan concatination

Kemudian, akan dicari nilai maksimal dan nama partikulatnya untuk mencari kategori ispu secara manual, dengan keterangan kategori (berdasarkan rentang nilai maksimal) ispu dikatakan:


*   baik: 0-50
*   sedang: 51-100
*   tidak sehat: 101-199
*   sangat tidak sehat: 200-299
*   berbahaya: >=300



```python
raw['maks'] = raw[['pm10','pm25','so2','co','o3','no2']].max(axis=1)
raw
```

```python
raw['nmaks'] = raw[['pm10','pm25','so2','co','o3','no2']].idxmax(axis=1)
#bisa dengan raw.iloc[:,1:7].idxmax(axis=1)
raw
```

```python
conditions = [
    (raw['maks'] >= 0) & (raw['maks'] <= 50),
    (raw['maks'] >= 51) & (raw['maks'] <= 100),
    (raw['maks'] >= 101) & (raw['maks'] <= 199),
    (raw['maks'] >= 200) & (raw['maks'] <= 299),
    (raw['maks'] >= 300)
    ]

values = ['baik', 'sedang', 'tidak sehat', 'sangat tidak sehat', 'berbahaya']

raw['kategori'] = np.select(conditions, values)

raw
```

```python
found=raw.iloc[:,7:10]
found
```

Presentase hari dengan kategori tertentu

```python
hari_colors = ['#0384fc', '#fc8003']

## fungsi labeling untuk menampilkan hari pada pie chart
def hari_labeling(values):
    def labeling(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%\n({v:d})'.format(p=pct,v=val)
    return labeling
```

```python
# Visualisasi persentase kategori hari

hari_labels = raw['kategori'].value_counts().index
hari_values = raw['kategori'].value_counts().values

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(hari_values,
        labels=hari_labels,
        counterclock=False,
        colors=hari_colors,
        autopct=hari_labeling(hari_values),
        shadow=True
         )

plt.title("Kategori Hari")
plt.show()
```

## **Data Outlier**

Menghilangkan data outlier untuk mempermudah prediksi

```python
Q1 = (raw[['pm10','pm25','so2','co','o3','no2']]).quantile(0.25)
Q3 = (raw[['pm10','pm25','so2','co','o3','no2']]).quantile(0.75)

IQR = Q3 - Q1
maximum = Q3 + (1.5*IQR)
print('Nilai Maximum dari masing-masing Variable adalah: ')
print(maximum)
minimum = Q1 - (1.5*IQR)
print('\nNilai Minimum dari masing-masing Variable adalah: ')
print(minimum)

more_than = raw[['pm10','pm25','so2','co','o3','no2']] > maximum
lower_than = raw[['pm10','pm25','so2','co','o3','no2']] < minimum
raw = raw.mask(more_than,maximum,axis=1)
raw = raw.mask(lower_than, minimum, axis=1)

print('\nPersebaran data setelah ditangani Outlier: ')
print(raw[['pm10','pm25','so2','co','o3','no2']].describe())
```
